## Building an interactive plot in steps

### Step 1: Hard-coding a value

Now we will use this data to create a figure with `imshow` where the colormap is "hard coded" as the "jet" colormap (not a great choice!).

Call this figure in the "usual way" with

```python
fig_1, ax_1 = plt.subplots()
```

And set the colormap's "hard coded" as the parameter `cmap_variable = 'jet'` (see the class notebook for more details).

**Make sure you have checked the naming of your variables ("fig_1", "ax_1", "cmap_variable").**

````python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
import matplotlib.pyplot as plt
fig_1, ax_1 = plt.subplots()
cmap_variable = 'jet'
im = ax_1.imshow(data, cmap = cmap_variable)
````

## Step 2: Parameter as function

Now create a function called `create_plot1` that takes the colormap in as a variable.

````python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
def create_plot1(map):
    fig_1, ax_1 = plt.subplots()
    cmap_variable = 'jet'
    im = ax_1.imshow(data, cmap = map)

create_plot1('jet')
````

## Step 2.5: Test out some values

Use your new function to test out some values.

````python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
create_plot1('Blues')
````

## Step 3

Using an `@ipywidgets.interact` decorator, add a dropdown menu to select the plot color with the options of ['jet', 'magma', 'Blues'].

Call this new, decorated function `create_plot2`.

**Make sure to call the function `create_plot2` and to use the correct order for your dropdown.**

When finished, you should see a plot with a dropdown that looks like this:

<div>

</div>

````python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
import ipywidgets
@ipywidgets.interact(style=['jet', 'magma', 'Blues'])

def create_plot2(style):
    fig_1, ax_1 = plt.subplots()
    cmap_variable = 'jet'
    im = ax_1.imshow(data, cmap = style)
````
